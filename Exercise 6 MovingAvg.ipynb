# Time Series - Moving Average Exercise 2

```python
import pandas as pd
from matplotlib import pyplot 
```

```python
df=pd.read_csv('AMZN.csv')
df.plot()
df
```

```python
pyplot.scatter(df['Date'],df['ClosingPrice'])
pyplot.show()
```

```python
PERIOD=200
df.ClosingPrice[-PERIOD:]
```

```python
df_endingfive=df[-PERIOD:]
df_endingfive
```

```python
df_endingfive.mean()
```

```python
df_endingfive['Volume'].sum()
```

```python
y=(df.ClosingPrice.rolling(window=200).mean())
```

```python
x=(df.ClosingPrice.rolling(window=50).mean())
```

```python
pyplot.plot(df.Date,df.ClosingPrice,label="Price")
pyplot.plot(df.Date,x,label="50 Day Moving Average",color="red")
pyplot.plot(df.Date,y,label="200 Day Moving Average")
pyplot.legend(loc='upper left')
pyplot.show()
```
